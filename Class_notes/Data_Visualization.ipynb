```python
# Normal curve

# importing matplotlib libraries

# import main matplotlit module
import matplotlib

# import pyplot sub-module from matplotlib
import matplotlib.pyplot as plt

# this line tells ipython notebooks to show all plots inline
%matplotlib inline

# Importing math library to use sqrt and other functions
import math

# Importing mlab module to use normal probability distribution function to construct a normally distributed variable
import matplotlib.mlab as mlab

# setting the mean and variance values
mean = 20
var_squared = 20.0
var = math.sqrt(var_squared)

# Generate 100 values of x from 0 to 40.
x = np.linspace(0, 40, 100)

# Using normpdf function to generate normally distributed y values corresponding to x values, mean and var of distribution
y = mlab.normpdf(x, mean, var)

# this line calls the figure method to manipulate the plot figure attributes like size of the plot
plt.figure(figsize=(20,8))

# Setting labels for X-axis and Y-axis
plt.xlabel('X')
plt.ylabel('Probability')

# plotting X and Y
dv_a = plt.plot(x, y)
```

```python
# scatterplot

# declaring empty lists
x_scat=[]
y_scat=[]

# loop to randomly generate 50 observations each for x and y
for i in range(0,50):
    x_scat.append(np.random.randint(1,10))
    y_scat.append(np.random.randint(5,200))

# Plotting a scatter plot with x_scat and y_scat lists
plt.scatter(x_scat,y_scat)

# Labels and titles
plt.xlabel('X values')
plt.ylabel('Y values')
plt.title('Random Scatter plot')
```

```python
# scatterplot example2

x_scat=[]
y_scat=[]

# Randomly picking 50 x values from the given range of (1,10)
for i in range(0,50):
    x_scat.append(np.random.randint(1,10))

# Defining y with a function, rather than randomly picking values from a range
for x in x_scat:
    y_scat.append(x**2)

# Plotting scatter    
plt.scatter(x_scat,y_scat)

# Labels and titles
plt.xlabel('X values')
plt.ylabel('Y values')
plt.title('Random Scatter plot')
```

```python
# subplot - dividing the plot canvas into a grid to show multiple plots side-by-side

# new empty list
z_scat=[]

# Randomly picking values for x and y
for i in range(0,50):
    x_scat.append(np.random.randint(1,10))
    y_scat.append(np.random.randint(5,200))

# Calculating z values using function    
for x in x_scat:
    z_scat.append(x**2) 

# Creating a subplot grid in the plot canvas, subplot(no. of rows, no. of columns, graph number)
plt.subplot(1,2,1)
plt.scatter(x_scat,y_scat)

plt.subplot(1,2,2)
plt.scatter(x_scat,z_scat)
```

```python
# distribution plot

# importing seaborn
import seaborn as sns

# randomly picking 100 values from normal distribution and plotting a distribution plot with histogram and distribution curve
x = np.random.normal(size=100)
sns.plt.ylabel('Probability') 
sns.plt.xlabel('X')
sns.plt.title('Distribution plot example')
y = sns.distplot(x)
```

```python
# KDE plot - Kernel Density Estimation plot

# loading iris dataset
iris = sns.load_dataset("iris")

# Plotting univariate and bivariate distributions
g = sns.PairGrid(iris)

# Univariate density estimation on diagonal
g.map_diag(sns.kdeplot,shade=True)

# Bivariate density estimation off the diagonal
g.map_offdiag(sns.kdeplot, cmap="Blues_d", n_levels=6,shade=True);
```

```python
# Pairplot

sns.pairplot(iris);
```

```python
# sinplot function
def sinplot(flip=1):
    x = np.linspace(0, 14, 100)
    for i in range(1, 7):
        plt.plot(x, y=np.sin(x + i * .5) * (7 - i) * flip)

# Example of plotting multiple lines/trends in a single plot
# Chips vs Salsa
# plt.plot(x, y=Sales of Salsa)
# Chips vs Guacomole
# plt.plot(x, y=Sales of Guac)
# Chips vs Ranch
# plt.plot(x, y=Sales of Ranch)

sinplot()
```

```python
# Single sin function plot

x = np.linspace(-np.pi, np.pi, 100)
plt.plot(x, np.sin(x))
plt.xlabel('Theta')
plt.ylabel('y=sin(x)')
plt.show()
```
