# Seaborn Styles & Context

```python
%matplotlib inline
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
np.random.seed(sum(map(ord, "aesthetics")))
```

```python
def sinplot(flip=1):
    x = np.linspace(0, 14, 100)
    for i in range(1, 7):
        plt.plot(x, np.sin(x + i * .5) * (7 - i) * flip)
```

```python
#Use sns.set to make it a seaborn plot
sns.set()
sinplot()
```

```python
#Changing the style to white grid. You could try darkgrid(default), dark, white and ticks
sns.set_style("darkgrid")
data = np.random.normal(size=(20, 6)) + np.arange(6) / 2
sns.boxplot(data=data);
```

```python
#Getting the ticks on x and y axes
sns.set_style("ticks")
sinplot()
```

```python
#Removing the spines on top and right
sinplot()
sns.despine()
```

```python
#You could offset the spines to get better rendering of the plot on the data
f, ax = plt.subplots()
sns.violinplot(data=data)
sns.despine(offset=10);
```

```python
#Trim parameter limits the spines range for aestheticity
f, ax = plt.subplots()
sns.violinplot(data=data)
sns.despine(offset=10, trim=True);
```

```python
#Try a box plot on a white grid
sns.set_style("whitegrid")
sns.boxplot(data=data, palette="deep")
sns.despine(left=True)
```

```python
#Temporarily set a style using a with statement and axes_style
sns.set_style("white")
with sns.axes_style("dark"):
    plt.subplot(211)
    sinplot()
plt.subplot(212)
sinplot(-1)
```

```python
#Call axes_style function to read all the style properties
sns.axes_style()
```

```python
#Overriding style parameters
sns.set_style("ticks", {"xtick.color": ".15"})
sinplot()
```

```python
#The four preset contexts, in order of relative size, are paper, notebook, talk, and poster. The notebook style is the default
#Set a poster context
sns.set_context("poster")
sinplot()
```

```python
#Also set style parameters using set_context
sns.set_context("notebook", font_scale=1.5, rc={"lines.linewidth": 1.5})
sinplot()
```
