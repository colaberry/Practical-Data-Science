```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
sns.set(color_codes=True)
#If True and palette is a seaborn palette
```

```python
np.random.seed(sum(map(ord, "regression")))
```

```python
#Load seaborn's tips dataset using load_dataset
tips = sns.load_dataset("tips")
tips.head()
```

```python
#regplot draws the scatterplot of two variables (bill & tips) and fits a linear regression line using y ~ f(x) and draws the line 
sns.regplot(x="total_bill", y="tip", data=tips);
```

```python
#lm plot also plots the linear regression line
sns.lmplot(x="total_bill", y="tip", data=tips);
```

regplot is different from lmplot that it is more flexible interms of data input - can be numpy arrays, pandas series or dataframe columns.

Let us use anscombe dataset

```python
#Load anscombe's dataset
#https://en.wikipedia.org/wiki/Anscombe%27s_quartet for more details about the dataset
anscombe = sns.load_dataset("anscombe")
anscombe.head()
```

```python
anscombe['dataset'].unique()
```

```python
#plot the 1st dataset in anscombe's using lmplot
sns.lmplot(x="x", y="y", data=anscombe.query("dataset == 'I'"),
           ci=None, scatter_kws={"s": 100});
```

```python
#second dataset - Note this is not a good linear model
sns.lmplot(x="x", y="y", data=anscombe.query("dataset == 'II'"),
           ci=None, scatter_kws={"s": 80});
```

```python
#In case of polynomial data, we can use the order parameter to plot higher order data
sns.lmplot(x="x", y="y", data=anscombe.query("dataset == 'II'"),
           order=2, ci=None, scatter_kws={"s": 80});
```

```python
#Outliers
sns.lmplot(x="x", y="y", data=anscombe.query("dataset == 'III'"),
           ci=None,  scatter_kws={"s": 80});
```

```python
#In case of such ouliers, we can use parameter robust=True
#This removes the outlier from the estimate function and plots the other data points better.
sns.lmplot(x="x", y="y", data=anscombe.query("dataset == 'III'"),
           robust=True, ci=None, scatter_kws={"s": 80});
```
