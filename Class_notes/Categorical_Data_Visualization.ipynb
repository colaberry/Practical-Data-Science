# Plotting Categorical Data

 1. swarmplot() and stripplot()
 2. boxplot() and violinplot()
 

```python
#import numpy,pandas,matplotlib
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
#import seaborn visualization library
import seaborn as sns
sns.set(style="whitegrid", color_codes=True)
np.random.seed(sum(map(ord, "categorical")))
```

```python
#load tips dataset
tips = sns.load_dataset("tips")
tips.head()
tips['time'].unique()
```

```python
#Draw a jointplot of total_bill Vs tip data
g = sns.jointplot(x="total_bill", y="tip", data=tips)
```

```python
# Stripplot on categorical variable shows the categories
#This is equivalent of scatterplot specifically if a variable is categorical
#For example the day variable is categorical
sns.stripplot(x="day", y="total_bill", data=tips);
```

```python
#User jitter=True to avoid overlapping of values
sns.stripplot(x="day", y="total_bill", data=tips, jitter=True);
```

```python
#Swarm plot gives the full distribution - a better visualizaton
sns.swarmplot(x="day", y="total_bill", data=tips);
```

```python
#use a hue to show a split on another categorical variable
sns.swarmplot(x="day", y="total_bill", hue="sex", data=tips)
```

```python
#If a categorical variable is numerical, it will be displayed in sorted order
sns.swarmplot(x="size", y="total_bill", data=tips);
```

```python
#it is sometimes better to plot the categorical variable in y axis, especially the labels for categories are long strings
sns.swarmplot(x="total_bill", y="day", hue="time", data=tips);
```

```python
#Box plot shows distribution within categories
sns.boxplot(x="day", y="total_bill", hue="time", data=tips);
```

```python
# A Good visualization using categorical variable is violin plot
#it gives the categories and also the Kernel density curve of other numerical variable within each category
sns.violinplot(x="total_bill", y="day", hue="time", data=tips);
```

```python
#Let us now plot with hue as sex variable
sns.violinplot(x="total_bill", y="day", hue="sex", data=tips);
```

```python
#Let us swap the axes and use split=True to have a more compact view.
#This is preferred espacially if we are dealing with binary categories. In this case sex has only 2 categories 
sns.violinplot(x="day", y="total_bill", hue="sex", data=tips, split=True,
               inner=None, palette={"Male": "b", "Female": "y"})
```

```python
#You could also combine violinplots with swarmplots to view the distribution
sns.violinplot(x="day", y="total_bill", data=tips, inner=None)
sns.swarmplot(x="day", y="total_bill", data=tips, color="w", alpha=.5);
```
